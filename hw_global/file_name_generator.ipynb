```python
# Generate the file list based on the provided pattern and the requested date range

base_pattern = "clmforc.GSWP3.c2011.0.5x0.5.TPQWL."
start_year = 1985
end_year = 2015

# Generate file names for each month of each year in the range
file_list = [f"{base_pattern}{year}-{month:02d}.nc" for year in range(start_year, end_year + 1) for month in range(1, 13)]

# Print the full list
for file_name in file_list:
    print(file_name)
```
